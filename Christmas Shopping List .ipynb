```python
#Import necessary Libraries
import requests
from bs4 import BeautifulSoup
import csv

# Define the URL
url = "https://www.macys.com/shop/gift-guide/holiday-gift-guide/gifts-by-recipient/christmas-gifts-for-kids?id=91556"

# Set headers to a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Sending the request to get the response
response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Extracting product names and prices
    product_data = []
    products_title = soup.find_all("div", class_="product-name medium")
    price_tag = soup.find_all("span", class_="discount price-reg")
    
    for name, price_tag in zip(products_title, price_tag):
        name = name.get_text(strip=True)
        price = float(price_tag.get_text(strip=True).replace("USD", "").replace(",", "").replace("Sale", "").replace('Your Choice','').replace("-",""))

 # Checking if price is valid and below $35
        if price and price <= 35:
            product_data.append({"name": name, "price": price})
    
    # Saving data to a CSV file
    with open("macy_products.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["name", "price"])
        writer.writeheader()
        writer.writerows(product_data)
    
    print("Data saved to macy_products.csv!")
else:
    print(f"Failed to retrieve webpage. Status code: {response.status_code}")
```
